```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

```

```python
df=pd.read_csv('../Data/Telco_customer_churn.csv')
```

```python
df_copy=df.copy()
```

```python
df.head()
```

```python
cols_to_remove = [
    "CustomerID",
    "Count",
    "Country",
    "State",
    "City",
    "Zip Code",
    "Lat Long",
    "Latitude",
    "Longitude",
    "Churn Label",
    "Churn Score",
    "CLTV",
    "Churn Reason"
]

df_copy = df_copy.drop(columns=cols_to_remove)
```

```python
df_copy.head()
```

```python
# select non-numerical columns
non_binary_cols = [

    "Online Security",
    "Online Backup",
    "Device Protection",
    "Tech Support",
    "Streaming TV",
    "Multiple Lines",
    "Streaming Movies"
]

# print unique values for each categorical column
for col in binary_cols:
    print(f"\n{col}")
    print(df_copy[col].unique())
```

```python
binary_cols = [
    "Partner",
    "Dependents",
    "Senior Citizen",
    "Phone Service",
    "Paperless Billing",
    
]

df_copy[binary_cols] = df_copy[binary_cols].replace({"Yes": 1, "No": 0})

df_copy["Gender"] = df_copy["Gender"].map({"Male": 1, "Female": 0})
```

```python
df_copy.head()
```

```python
df_copy = pd.get_dummies(
    df_copy,
    columns=[
        
        "Contract",
        "Payment Method",
        "Internet Service",
        "Online Security",
        "Online Backup",
        "Device Protection",
        "Tech Support",
        "Streaming TV",
        "Multiple Lines",
        "Streaming Movies"
    ],
    drop_first=True
)
```

```python
df_copy.head(5)
```

```python
bool_cols = df_copy.select_dtypes(include="bool").columns
df_copy[bool_cols] = df_copy[bool_cols].astype(int)
```

```python
df_copy.head()
```

```python
df_copy.dtypes
```

```python
df_copy.isna().sum().sort_values(ascending=False)
```

```python

obj_cols = df_copy.select_dtypes(include="object").columns

df_copy[obj_cols] = df_copy[obj_cols].apply(
    lambda x: pd.to_numeric(x, errors="coerce")
)
```

```python
df_copy[obj_cols] = df_copy[obj_cols].fillna(
    df_copy[obj_cols].median()
)
```
